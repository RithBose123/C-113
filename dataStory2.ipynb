```python
import pandas as pd
import plotly.express as px
df=pd.read_csv("data.csv")
fig=px.scatter(df,y="quant_saved",color="wealthy")
fig.show()
```

```python
import csv
with open("data.csv",newline="")as f:
  reader=csv.reader(f)
  savedData=list(reader)
savedData.pop(0)
totalEntries=len(savedData)
females=0
for i in savedData:
  if int(i[1])==1:
    females+=1
import plotly.graph_objects as go
fig=go.Figure(go.Bar(x=["female","male"],y=[females,(totalEntries-females)]))
fig.show()
```

```python
import statistics as st
allSavings=[]
for data in savedData:
  allSavings.append(float(data[0]))
print(f"mean of savings- {st.mean(allSavings)}")
print(f"median of savings- {st.median(allSavings)}")
print(f"mode of savings- {st.mode(allSavings)}")
print(f"Standard Deviation of savings- {st.stdev(allSavings)}")
```

```python
maleSaving=[]
femaleSaving=[]
for i in savedData:
  if int(i[1])==1:
    femaleSaving.append(float(i[0]))
  else:
    maleSaving.append(float(i[0]))
print(f"the mean of savings of females is {st.mean(femaleSaving)}")
print(f"the mode of savings of females is {st.mode(femaleSaving)}")
print(f"the median of savings of females is {st.median(femaleSaving)}")
print(f"the standard deviation of savings of females is {st.stdev(femaleSaving)}")
print("\n")
print(f"the mean of savings of males is {st.mean(maleSaving)}")
print(f"the mode of savings of males is {st.mode(maleSaving)}")
print(f"the median of savings of males is {st.median(maleSaving)}")
print(f"the standard deviation of savings of males is {st.stdev(maleSaving)}")
```

```python
import numpy
wealth=[]
savings=[]
for i in savedData:
  wealth.append(int(i[3]))
  savings.append(float(i[0]))
correlation=numpy.corrcoef(wealth,savings)
print("the correlation is ",correlation[0,1])
```

```python
import plotly.figure_factory as ff

fig=ff.create_distplot([df["quant_saved"]],["savings"],show_hist=False)
fig.show()
```

```python
q1=df["quant_saved"].quantile(0.25)
# InterQuartileRange
q3=df["quant_saved"].quantile(0.75)
iqr=q3-q1
lowerWhisker=q1-1.5*iqr
upperWhisker=q3+1.5*iqr
print(f"q1- {q1}")
print(f"q3- {q3}")
print(f"iqr is - {iqr}")
print(f"lower whisker- {lowerWhisker}")
print(f"upper whisker- {upperWhisker}")
newdf=df[df["quant_saved"]<upperWhisker]
```

```python
# import statistics as st
saving=newdf["quant_saved"].tolist()

print(f"mean of savings- {st.mean(allSavings)}")
print(f"median of savings- {st.median(allSavings)}")
print(f"mode of savings- {st.mode(allSavings)}")
print(f"Standard Deviation of savings- {st.stdev(allSavings)}")
fig=ff.create_distplot([newdf["quant_saved"].tolist()],["quantity saved"],show_hist=False)
fig.show()
```
